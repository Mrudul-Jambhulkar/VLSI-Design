```python
import numpy as np

# Data
loads = np.array([1, 2, 3, 4, 5, 6, 7, 8])
delay = np.array([3.942162e-11, 5.180302e-11, 6.377595e-11, 7.563500e-11,8.744390e-11, 9.925118e-11, 1.110666e-10 , 1.228934e-10])



# Fit a linear regression line
slope, intercept = np.polyfit(loads, delay, 1)



import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(loads, delay, label="Data")
plt.plot(loads, slope * loads + intercept, color='red', label="Fitted Line")

# Add labels and legend
plt.xlabel("Number of Loads")
plt.ylabel("Delay")
plt.legend()

# Show the plot
plt.grid()
plt.show()



tau = slope
pinv = intercept
print(f"tau: {tau}")

print(f"pinv: {pinv}")

print(f"gamma: {1.4/0.49}")
```
